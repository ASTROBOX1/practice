## ex 1

```python
total=0
for x in range(1,5):
    score=int(input(f"Enter the score for subject {x} : "))
    total+=score

if total >=90:
    grad ='A'
elif  80 < total <90:
    grad ='B'
elif 70 < total <80:
    grad ='C'
else: grad ='D'    

print(f"total Score : {total}")
print(f"grad : {grad}")
```

## ex 6

```python
result=1
num = int(input('enter your number : '))
if num >0:
    for x in range(1,num+1):
        result*=x
    print(f"the factorial is {result}")    
elif num==0:
    result =1
    print(f"the factorial is {result}")
else : 
    print('invalid number ')            
```

## ex 12

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
tuple3 = (7, 8, 9)
length = len(tuple1)
result = []
for i in range(length):
    total = tuple1[i] + tuple2[i] + tuple3[i]
    result.append(total)
print("Element-wise sum of tuples:", result)
```
