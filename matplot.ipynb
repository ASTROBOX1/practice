```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

sns.barplot(x=haiti.index, y=haiti.values, color='r', alpha=0.5, label='Bar Chart')

sns.lineplot(x=range(len(haiti)), y=haiti.values, color='darkred', marker='o', linestyle='--', linewidth=2, label='Trend Line')


plt.title('Haiti Immigration Trend', fontsize=15)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Numbers', fontsize=12)
plt.xticks(rotation=45) 
plt.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

ax = sns.barplot(x=haiti.index, y=haiti.values, color='r', alpha=0.8)

sns.lineplot(x=range(len(haiti)), y=haiti.values, marker='o', linestyle='--')

ax.bar_label(ax.containers[0], padding=3, fontsize=10)

plt.title('Haiti Immigration Data with Values', fontsize=16)
plt.xlabel('Years', fontsize=12)
plt.ylabel('Numbers', fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

```python
df_coun=df.groupby('Continent').sum()
```

```python
sizes = df_coun['Total']
labels = df_coun.index

plt.figure(figsize=(5, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')

plt.show()
```
