```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, ExtraTreesClassifier
```

```python
models = {

    # ============================================
    # Decision Tree Regressor
    # ============================================
    "Decision Tree": DecisionTreeClassifier(
        # Splitting criterion
        criterion='entropy', #{"squared_error", "friedman_mse", "absolute_error", "poisson"}
        
        # Tree depth
        max_depth=5,

        # Minimum samples to split node
        min_samples_split=10,

        # Minimum samples in leaf
        min_samples_leaf=5,

        # Minimum improvement to split
        min_impurity_decrease=0.01,

        # Pruning parameter
        ccp_alpha=0.1,

        random_state=100
    ),


    # ============================================
    # Random Forest Regressor
    # ============================================
    "Random Forest": RandomForestClassifier(

        # Number of trees
        n_estimators=50,

        # Splitting criterion
        criterion='entropy', #{"squared_error", "friedman_mse", "absolute_error", "poisson"}
        
        # Tree depth
        max_depth=None,

        # Features per split
        max_features="sqrt",

        random_state=100
    ),


    # ============================================
    # Bagging Regressor
    # ============================================
    "Bagging": BaggingClassifier(

        estimator=DecisionTreeClassifier(
            criterion='entropy',
            max_depth=5,
            min_samples_split=10,
            min_samples_leaf=5,
            min_impurity_decrease=0.01,
            ccp_alpha=0.1,
            random_state=100
        ),

        n_estimators=50,


        bootstrap=True,

        random_state=100
    ),


    # ============================================
    # Extra Trees Regressor
    # ============================================
    "Extra Trees": ExtraTreesClassifier(
        criterion='entropy', #{"squared_error", "friedman_mse", "absolute_error", "poisson"}
        n_estimators=50,
        max_depth=None,
        max_features="sqrt",

        random_state=100
    )
}
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
accuracy=[]
for model_name , model in models.items():
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)

    accu_train=accuracy_score(y_train,y_pred_train)
    accu_test=accuracy_score(y_test,y_pred_test)

    
    f1_train=f1_score(y_train,y_pred_train,average='micro')
    f1_test=f1_score(y_test,y_pred_test,average='micro')

    
    accuracy.append([accu_train,
            accu_test,
            f1_train,
            f1_test]
            )

    


```

```python
accuracy=pd.DataFrame(accuracy,columns=['accu_train',
            'accu_test',
            'f1_train',
            'f1_test'])
```
